```python
from sklearn.model_selection import train_test_split
```

```python
patients.columns
```

```python
X=patients[['Age', 'Gender', 'Total_Bilirubin', 'Direct_Bilirubin',
       'Alkaline_Phosphotase', 'Alamine_Aminotransferase',
       'Aspartate_Aminotransferase', 'Total_Protiens', 'Albumin',
       'Albumin_and_Globulin_Ratio']]
y=patients['Dataset']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=123)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
```

```python
ranmodel = RandomForestClassifier(n_estimators = 20)
ranmodel.fit(X_train, y_train)
```

```python
# accuracy score on the training data
X_train_prediction_rf = ranmodel.predict(X_train)
training_data_accuracy_rf = accuracy_score(X_train_prediction_rf, y_train)
```

```python
print('Accuracy score of the training data for random forest model is : ', training_data_accuracy_rf)
```

```python
confusion_matrix(y_test, ranmodel.predict(X_test))
```

```python
input_data = (17,1,0.9,0.3,202,22,19,7.4,4.1,1.20)
#input_data = (65,0,0.7,0.1,187,16,18,6.8,3.3,0.90)

# changing the input_data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the array as we are predicting for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = ranmodel.predict(input_data_reshaped)
print(prediction)

if (prediction[0] == 0):
  print('The person is not having Chronic Kidney Disease')
else:
  print('The person is having Chronic Kidney Disease')
```
