```python
le=LabelEncoder()
dummy=le.fit_transform(train_df['prognosis'])
y_train=pd.DataFrame(dummy)
```

```python
dummy=LabelEncoder().fit_transform(test_df['prognosis'])
y_test=pd.DataFrame(dummy)
y_test.head()
len(y_test)
```

```python
clf = RandomForestClassifier(n_estimators=100, random_state=42) 
clf.fit(x_train, y_train)
```

```python
rf_predictions = clf.predict(x_test)
print(len(rf_predictions))
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, accuracy_score, classification_report
mae = mean_absolute_error(y_test, rf_predictions)
mse = mean_squared_error(y_test, rf_predictions)
r2 = r2_score(y_test, rf_predictions)

print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("R-squared:", r2)
```

```python
training_data_accuracy = accuracy_score(clf.predict(x_train), y_train)
testing_data_accuracy =  accuracy_score(clf.predict(x_test), y_test)
```

```python
print('Accuracy score of the training data for random forest model is : ', training_data_accuracy)
print('Accuracy score of the testing data for random forest model is : ', testing_data_accuracy)
```

```python
confusion_matrix(y_test, clf.predict(x_test))
```
